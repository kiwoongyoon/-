# Batch normalization, dropout에 따른 모델의 성능 변화 이해
>Fasion MNIST 데이터에서 drop out rate를 0.2 , 0.5, 0.8로 바꿔가며 정확도가 변화하는 양상을 관찰해본다.

```python
import os, random
import keras
from keras.datasets import fashion_mnist
from keras.models import Sequential
from keras.layers import Dense, BatchNormalization, Dropout
from keras.utils import np_utils
import numpy as np
import tensorflow as tf
os.path.expanduser = lambda path: './'

batch_size = 128
num_classes = 10
epochs = 60

# the data, split between train and test sets
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

x_train = x_train.reshape(60000, 784)
x_test = x_test.reshape(10000, 784)
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')

# convert class vectors to binary class matrices
y_train = keras.utils.np_utils.to_categorical(y_train, num_classes)
y_test = keras.utils.np_utils.to_categorical(y_test, num_classes)

# for reproducibility
import random, os
os.environ['PYTHONHASHSEED']='0'
random.seed(123)
np.random.seed(123)
tf.random.set_seed(123)
sess = tf.compat.v1.Session(config=tf.compat.v1.ConfigProto(intra_op_parallelism_threads=1,    inter_op_parallelism_threads=1,
                                                               allow_soft_placement=True, device_count = {'CPU': 1}))
from tensorflow.python.keras import backend as K
K.set_session(sess)



kernel_initializer='glorot_uniform'
activation_function = 'relu'

with tf.device('/cpu:0'):
    model = Sequential()
    model.add(Dense(512, activation='relu', input_shape=(784,)))
    model.add(Dropout(0.2))
    #model.add(BatchNormalization())
    model.add(Dense(512, activation='relu'))
    model.add(Dropout(0.2))
    #model.add(BatchNormalization())
    model.add(Dense(num_classes, activation='softmax'))

    model.summary()

    model.compile(loss='categorical_crossentropy',
                  optimizer='sgd',
                  metrics=['accuracy'])

history = model.fit(x_train, y_train,
                    batch_size=batch_size,
                    epochs=epochs,
                    validation_split=0.8)


metrics = model.evaluate(x_test, y_test) #returns loss and accuracy
print(metrics[1])
print(f'Accuracy: {metrics[1]*100:.2f}%\n')
```

 dropout을 적용하는 이유는 딥러닝 학습을 하면서 생기는 문제인 overfitting을 해소하기 위함이다. 딥러닝의 hidden layer의 일부 유닛이 동작하지 않게 하여 overfitting을 막는 것이다. 대체로 dropout의 값이 0.3에서 0으로 줄어들면 너무 많이 drop하여 에러의 발생확률이 더 커진다. 또한, dropout의 값이 0.8에서 1로 커지는 과정에서는 drop을 너무 적게 적용시켜서 dropout을 사용하는 의미가 퇴색된다. 그리하여 적절한 dropout의 확률은 0.5가 된다고 알려져있다. 
 Fasion MNIST의 데이터를 이용한 dropout에 따른 정확도 측정에서 drop이 각각 0.2,0.5,0.8일 경우 정확도가 84.38%, 85.84%, 84.82%로 0.5인 경우에 가장 정확도가 높았고 0.2와 0.8인경우 drop이 0.5인 경우보다 정확도가 1%이상 떨어졌다. 결론적으로 적절한 dropout 값을 찾는 것이 딥러닝 학습에서 중요하고 대체로 0.5 부근의 drop 확률이 정확성이 높다는 것을 알게 되었다.
